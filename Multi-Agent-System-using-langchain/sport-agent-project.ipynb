
### 2. Import Graph & Visualize Flow
#### Visualizing the graph helps verify the 'Gatekeeper -> Scout -> Analyst' flow.

```python

#import graph
from graph import app
```


 ### 3. Run Analysis (Valid Input)
#### This will trigger the full workflow: Gatekeeper -> Scout -> Analyst.

```python

inputs = {"entity_name": "Hardik Pandya"}
print(f"--- Starting Analysis for {inputs['entity_name']} ---")

for output in app.stream(inputs):
    for key, value in output.items():
        print(f"\n[Node: {key}]")
        # Print specific results if available
        if "scout_data" in value:
            print(f"Scout found data: {value['scout_data'][:200]}...") 
        if "analyst_report" in value:
            print(f"Analyst Report:\n{value['analyst_report']}")

```

 ### 4. Run Analysis (Invalid Input)
#### This demonstrates the Gatekeeper stopping the workflow.

```python

inputs_invalid = {"entity_name": "Write a poem about rain"}
print("\n--- Testing Invalid Input ---")

result = app.invoke(inputs_invalid)
print(f"Is Valid: {result.get('is_valid')}")
print(f"System Response: {result.get('analyst_report')}")
```
